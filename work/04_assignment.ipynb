```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

%config InlineBackend.figure_format ='retina'
```

```python
df = pd.read_csv("../work/mcdonalds")
df = df[['Category', 'Item', 'Calories']]
df = df[df['Category'].str.contains("Breakfast")]
df.head()
```

```python
plotdata = pd.DataFrame(
    {"Calories": [300, 250, 370, 450, 400]}, 
    index=["Egg McMuffind", "Egg White Delight", "Sausage McMuffin", "Sausage McMuffin with Egg", "Sausage McMuffin with Egg Whites"])
p1=plotdata.plot(kind="bar")
plt.title("Calories in McDonalds Breakfast Items")
plt.xlabel("Breakfast Items")
plt.ylabel("Calories")
```

```python
#5:This is a bar graph that shows the calories in popular McDonalds breakfast items.
```

```python
#6:I chose this visualization because it provides a simple way for customers to see what products have more or less calories. If a customer is looking to limit their calories theey can easily see the amount of calories in an item and which products have more or less calories.
```
